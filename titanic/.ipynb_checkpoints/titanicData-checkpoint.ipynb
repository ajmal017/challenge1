```python
import pandas as pd

df = pd.read_csv("train.csv")
df.head()
```

```python
dfRevCols = df[df.columns[::-1]]
dfRevCols.head()
```

```python
dfEveryOtherCol = df[df.columns[::2]]
dfEveryOtherCol.head()
```

```python
dfEveryThirdRev = df[df.columns[::3]]
dfEveryThirdRev.head()
```

```python
from pandas.api.types import is_string_dtype

def rev(string):
    return string[::-1]

strCols = [col for col in dfEveryThirdRev.columns if is_string_dtype(dfEveryThirdRev[col])]

for col in strCols:
    dfEveryThirdRev.loc[:,col] = dfEveryThirdRev[col].map(rev)

dfEveryThirdRev.head()
```
